```python
from sympy import *
from dtumathtools import *
init_printing()

r, theta, B_rem, p = symbols('r theta B_rem p',real=True)
V = lambda p: B_rem * Matrix([cos(p*theta), sin(p*theta)])
V(p)
```

```python
dtuplot.plot_vector(V(-5), (B_rem,-1,1),(theta,-1,1))
```

```python
# Define symbols
r, theta, B_rem, p = symbols('r theta B_rem p',real=True)

e_r = Matrix([cos(theta), sin(theta)])
e_theta = Matrix([-sin(theta), cos(theta)])

V = lambda p: (B_rem * cos(p*theta)) * e_r + (B_rem * sin(p*theta)) * e_theta
V(p)
dtuplot.plot_vector(V(-2), (B_rem,-5,5),(theta,-5,5))
```
